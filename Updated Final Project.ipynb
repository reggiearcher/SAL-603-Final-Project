```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
from pybaseball import lahman
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
import seaborn as sns
```

```python
weekly_rate = pd.read_csv('Weekly_IR.csv')
injury_classification = pd.read_csv('Injury_classification.csv')
weekly_rate['Injury_Rate'] = (weekly_rate['TL_Injuries'] / weekly_rate['Exposure']) * 1000
weekly_rate['NC-ST_IR'] = (weekly_rate['NC-ST'] / weekly_rate['Exposure']) * 1000
weekly_rate['C-ST_IR'] = (weekly_rate['C-ST'] / weekly_rate['Exposure']) * 1000
weekly_rate['M/L_IR'] = (weekly_rate['M/L'] / weekly_rate['Exposure']) * 1000
weekly_rate['F/D/S_IR'] = (weekly_rate['F/D/S'] / weekly_rate['Exposure']) * 1000
weekly_rate['Con_IR'] = (weekly_rate['Con'] / weekly_rate['Exposure']) * 1000
weekly_rate
injury_classification
```

```python
# Using the grouped by function to find the descriptive statistics of the injury classification dataframe. 
# Using the mean function we calcuated the average days missed by classification 
grouped_classifications = injury_classification.groupby('Classification')[['Days_Missed']].mean()
grouped_classifications
```

```python
# Mapping the frequency distribution of injuries suffered during one season of competition. 
# Only time loss injuries were used for analysis 

counts = injury_classification['Classification'].value_counts()
plt.figure(figsize=(10, 6)) 
plt.bar(counts.index, counts.values, color='green')
plt.xlabel('Classifications')
plt.ylabel('Counts')
plt.title('Injury Classification')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('Injury_Classification_Bar_Graph.pdf')
```

```python
# Obtaining measures of central tendency between the two time in seasons 
Preseason = weekly_rate['NC-ST_IR'].iloc[0:4]  # Data from index 0 to 3
Regular_Season = weekly_rate['NC-ST_IR'].iloc[4:17]  # Data from index 4 to 16
# Measures of central tendency between each group 
Preseason_mean = Preseason.mean()
Preseason_median = Preseason.median()
Preseason_std = Preseason.std()
Regular_Season_mean = Preseason.mean()
Regular_Season_median = Preseason.median()
Regular_Season_std = Regular_Season.std()
print("Preseason")
print("Mean:", Preseason_mean)
print("Median:", Preseason_median)
print("Standard Deviation:", Preseason_std)
print("Regular Season")
print("Mean:", Regular_Season_mean)
print("Median:", Regular_Season_median)
print("Standard Deviation:", Regular_Season_std)
```

```python
#Created a weekly injury count bar graph
# by adding dataframe_ax and saving to this we can export the graph. Then we set the figure object to .getfigure()
# Then we used the .savefig() to export the graph
weekly_rate.reset_index(inplace=True)
weekly_rate.set_index('Week', inplace=True)
weekly_rate_ax = weekly_rate[['NC-ST']].plot.bar()
header_text = "Weekly Non-Contact Soft-Tissue Injury Count"
weekly_rate_ax.set_title(header_text, fontsize=16)
plt.tight_layout()
weekly_rate_ax_figure = weekly_rate_ax.get_figure()
weekly_rate_ax_figure.savefig('Weekly_Injury_Count_By_Classification.pdf')
```

```python
# Below we used the weekly rate dataframe with specfic columns to plot a stacked bar graph 
# Only difference from above is that we used "stacked= True" within the function
weekly_rate.reset_index(inplace=True)
weekly_rate.set_index('Week', inplace=True)
combined_graph_ax = weekly_rate[['NC-ST','C-ST','Con']].plot.bar(stacked=True)
header_text = "Top 3 Injuries Classifications"
combined_graph_ax.set_title(header_text, fontsize=14)
combined_graph_ax_figure = combined_graph_ax.get_figure()
combined_graph_ax_figure.subplots_adjust(bottom=0.2)
plt.tight_layout()
combined_graph_ax_figure.savefig('Top 3 Injuries By Classification.pdf')
```

```python
#Below we created a boxplot to visualize days missed by classification 
Severity_ax = sns.catplot(x='Days_Missed', y='Classification', kind='box', data=injury_classification)
Severity_ax.set_titles("Days Missed By Classification", fontsize=14)
Severity_ax.savefig("Days_Missed_By_Classification.pdf")
plt.tight_layout()
```
